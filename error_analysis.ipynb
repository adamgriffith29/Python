```python
import matplotlib.pyplot as plt
import numpy as np
t, d, dy = np.loadtxt("Prob3.txt",
                      skiprows=1 , unpack=True)

plt.plot(t, d, 'ro')
plt.errorbar(t, d, xerr=0.001, yerr=dy, ecolor='blue')
plt.xlabel('time (t)')
plt.ylabel('y(t)')
plt.title('Velocity vs time')

plt.show()
```
