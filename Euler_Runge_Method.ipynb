```python
import numpy as np
import matplotlib.pyplot as plt
#Define Basic variables
x0 = 0
y0 = 1
xf = 10
n = 101
deltax = (xf-x0)/n
x = np.linspace(x0,xf,n)
y = np.zeros(n)
y[0] = y0
for i in range(1,n):
 y[i] = deltax*(-y[i-1] + np.sin(x[i-1])) + y[i-1]
plt.plot(x,y, 'o')
plt.xlabel(" x")
plt.ylabel(" y")
plt.title("Approximate Solution with Eulerâ€™s Method")
plt .show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *


def fun(t,y):
    fun =3*t**2
    return fun


#set step size
n=101
t0 = -1            
tmax = 1 
dt= (tmax-t0)/(n)     #Step size

#define arrays 
yrk= zeros(n,float)    # Solution with Runge-Kutta
ya= zeros(n,float)    # Solution analytical
t= zeros(n,float)    # time

#set initial condition 
yrk[0]=-1 #ic
ya[0]=-1 #ic
t[0]=t0 #ic

for i in range(1,n):

    k1 = dt*fun(t[i-1],yrk[i-1])
    k2 = dt*fun(t[i-1]+0.5*dt, yrk[i-1]+ 0.5*k1)
    k3 = dt*fun(t[i-1]+0.5*dt, yrk[i-1]+ 0.5*k2)
    k4 = dt*fun(t[i-1]+dt, yrk[i-1]+k3)
    #combine
    yrk[i] = yrk[i-1] + (k1+2*k2+2*k3+k4)/6

    t[i] = t[i-1] + dt
    #analytical
    ya[i] = t[i]**3

    

plt.plot(t, yrk, 'go', label='Runge-Kutta Method')
plt.plot(t, ya, 'k', label='Analytical')    
plt.legend()
plt.show()
    
```
