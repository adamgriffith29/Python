```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('NVIDIA_Stock_Volatility_2014_2024.csv')
data.head()
print(data.describe())
print(data.info)
data['Date'] = pd.to_datetime(data['Date'])
figure = px.line(data_frame = data,
                    x = 'Date', 
                    y = 'Daily_Return' , 
                    labels = {'Daily_Return':'Daily Return'},
                    title = 'Daily Return of NVIDIA Stock')
figure.show()
```
