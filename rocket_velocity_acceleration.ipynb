```python
import numpy as np
import matplotlib.pyplot as plt

t, a = np.loadtxt("rocketdata.dat", skiprows=1, unpack = True)
plt.subplot(3,1,1)
plt.plot(t,a,'k-')
plt.xlabel('t')
plt.ylabel('a(t)')
plt.title('acceleration v time')

#to get fromm acceleration to velocity, integration is required

x0 = 0
y0 = 1
xf = 15
n = 101
h = (xf-x0)/(n-1)
x = np.linspace(x0,xf,n)
y = np.zeros(n)
y[0] = y0
for i in range(1,n):
    y[i] = h*(-y[i-1] + (a[i-1])) + y[i-1]

plt.subplot(2,1,2)
plt.plot(x,y, 'o')
plt.xlabel("t")
plt.ylabel("v(t)")
plt.title("velocity v time")
plt.show()
```
