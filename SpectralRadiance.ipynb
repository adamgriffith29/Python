```python
import numpy as np
import matplotlib.pyplot as plt
x, y  = np.loadtxt("SpectralRadianceGreen.txt",
                  skiprows=1, unpack = True)
plt.plot(x, y, 'g-' , label = 'Green')
plt.title('Spectral Radiance for Green')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Spectral Radiance(W*m^-2 - SR)')

```
